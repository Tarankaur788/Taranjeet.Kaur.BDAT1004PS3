# Question 1

```python
import pandas as pd

# Step 3: Assign the dataset to a variable called 'users'
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, sep="|")  # Load the dataset using pipe ('|') as separator

# Step 4: Discover what is the mean age per occupation
mean_age_per_occupation = users.groupby('occupation')['age'].mean()

# Step 5: Discover the Male ratio per occupation and sort it from most to the least
male_ratio_per_occupation = users[users['gender'] == 'M'].groupby('occupation')['gender'].count() / users.groupby('occupation')['gender'].count()
male_ratio_per_occupation_sorted = male_ratio_per_occupation.sort_values(ascending=False)

# Step 6: For each occupation, calculate the minimum and maximum ages
min_max_age_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])

# Step 7: For each combination of occupation and sex, calculate the mean age
mean_age_per_occupation_gender = users.groupby(['occupation', 'gender'])['age'].mean()

# Step 8: For each occupation, present the percentage of women and men
gender_percentage_per_occupation = users.groupby('occupation')['gender'].value_counts(normalize=True).unstack(fill_value=0) * 100

# Displaying the results
print("Mean age per occupation:")
print(mean_age_per_occupation)

print("\nMale ratio per occupation (sorted from most to least):")
print(male_ratio_per_occupation_sorted)

print("\nMin and Max age per occupation:")
print(min_max_age_per_occupation)

print("\nMean age per occupation and gender:")
print(mean_age_per_occupation_gender)

print("\nGender percentage per occupation:")
print(gender_percentage_per_occupation)
```

```python
import pandas as pd

# Step 2: Import the dataset from the provided URL
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv"
euro12 = pd.read_csv(url)

# Step 3: The dataset is already assigned to the variable 'euro12' when we load it

# Step 4: Select only the Goal column
goal_column = euro12['Goals']

# Step 5: How many teams participated in Euro 2012?
teams_count = euro12.shape[0]

# Step 6: What is the number of columns in the dataset?
num_columns = euro12.shape[1]

# Step 7: View only the columns 'Team', 'Yellow Cards' and 'Red Cards'
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]

# Step 8: Sort the teams by 'Red Cards', then by 'Yellow Cards'
sorted_teams = euro12.sort_values(by=['Red Cards', 'Yellow Cards'], ascending=[False, False])

# Step 9: Calculate the mean Yellow Cards given per Team
mean_yellow_cards = euro12['Yellow Cards'].mean()

# Step 10: Filter teams that scored more than 6 goals
teams_more_than_6_goals = euro12[euro12['Goals'] > 6]

# Step 11: Select the teams that start with G
teams_start_with_G = euro12[euro12['Team'].str.startswith('G')]

# Step 12: Select the first 7 columns
first_7_columns = euro12.iloc[:, :7]

# Step 13: Select all columns except the last 3
all_columns_except_last_3 = euro12.iloc[:, :-3]

# Step 14: Present only the Shooting Accuracy from England, Italy, and Russia
shooting_accuracy = euro12[euro12['Team'].isin(['England', 'Italy', 'Russia'])][['Team', 'Shooting Accuracy']]

# Display the results
print("\nGoal column:")
print(goal_column.head())  # Display first few rows

print("\nNumber of teams that participated in Euro 2012:", teams_count)

print("\nNumber of columns in the dataset:", num_columns)

print("\nDiscipline DataFrame (Team, Yellow Cards, Red Cards):")
print(discipline.head())

print("\nTeams sorted by Red Cards and Yellow Cards:")
print(sorted_teams[['Team', 'Red Cards', 'Yellow Cards']].head())

print("\nMean Yellow Cards per Team:", mean_yellow_cards)

print("\nTeams that scored more than 6 goals:")
print(teams_more_than_6_goals[['Team', 'Goals']])

print("\nTeams that start with 'G':")
print(teams_start_with_G[['Team']])

print("\nFirst 7 columns:")
print(first_7_columns.head())

print("\nAll columns except the last 3:")
print(all_columns_except_last_3.head())

print("\nShooting Accuracy from England, Italy, and Russia:")
print(shooting_accuracy)
```

```python
import pandas as pd
import numpy as np

# Step 2: Create 3 different Series, each of length 100
# Series 1: Random numbers from 1 to 4 (representing the number of bedrooms)
bedrooms = pd.Series(np.random.randint(1, 5, size=100))

# Series 2: Random numbers from 1 to 3 (representing the number of bathrooms)
bathrooms = pd.Series(np.random.randint(1, 4, size=100))

# Series 3: Random numbers from 10,000 to 30,000 (representing the price per square meter)
price_sqr_meter = pd.Series(np.random.randint(10000, 30001, size=100))

# Step 3: Create a DataFrame by joining the Series by column
housing_df = pd.DataFrame({
    'bedrs': bedrooms,
    'bathrs': bathrooms,
    'price_sqr_meter': price_sqr_meter
})

# Step 4: Change the name of the columns to bedrs, bathrs, price_sqr_meter (already done above)

# Step 5: Create a one column DataFrame with the values of the 3 Series and assign it to 'bigcolumn'
bigcolumn = pd.DataFrame({
    'bigcolumn': pd.concat([bedrooms, bathrooms, price_sqr_meter], ignore_index=True)
})

# Step 6: Check if the DataFrame goes only until index 99
print("Does the DataFrame go only until index 99?")
print(housing_df.shape)  # This should show (100, 3), indicating 100 rows and 3 columns

# Step 7: Reindex the DataFrame so it goes from 0 to 299
# We want to reindex `bigcolumn` to have indices from 0 to 299
bigcolumn_reindexed = bigcolumn.reset_index(drop=True)

# Show the results
print("\nHousing DataFrame:")
print(housing_df.head())  # Display first 5 rows of the housing DataFrame

print("\nBigcolumn DataFrame with combined data:")
print(bigcolumn.head())  # Display first 5 rows of bigcolumn

print("\nReindexed Bigcolumn DataFrame (0 to 299):")
print(bigcolumn_reindexed.head())  # Display first 5 rows of reindexed bigcolumn
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for marriages and divorces per capita in the U.S. (hypothetical data for 1900, 1950, and 2000)
years = ['1900', '1950', '2000']

# Hypothetical values for marriages and divorces per capita (per 1,000 people)
marriages_per_capita = [7.5, 8.0, 7.0]  # Example data
divorces_per_capita = [0.5, 1.5, 4.0]   # Example data

# Set the positions for the bars
x = np.arange(len(years))

# Bar width
width = 0.35

# Create the plot
plt.figure(figsize=(10, 6))

# Plot bars for marriages and divorces
plt.bar(x - width/2, marriages_per_capita, width, label='Marriages per Capita', color='b')
plt.bar(x + width/2, divorces_per_capita, width, label='Divorces per Capita', color='r')

# Add title and labels
plt.title('Marriages and Divorces per Capita in the U.S. (1900, 1950, and 2000)')
plt.xlabel('Year')
plt.ylabel('Per Capita (per 1,000 people)')

# Add x-tick labels
plt.xticks(x, years)

# Show legend
plt.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Hypothetical data: actors and their respective kill counts in movies
actors = ['Bruce Willis', 'Samuel L. Jackson', 'Arnold Schwarzenegger', 'Tom Cruise', 'Dwayne Johnson']
kill_counts = [1000, 900, 800, 750, 600]

# Sort the actors by kill count (descending order)
sorted_indices = sorted(range(len(kill_counts)), key=lambda i: kill_counts[i], reverse=True)
sorted_actors = [actors[i] for i in sorted_indices]
sorted_kill_counts = [kill_counts[i] for i in sorted_indices]

# Create the horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(sorted_actors, sorted_kill_counts, color='green')

# Add labels and title
plt.xlabel('Kill Count')
plt.ylabel('Actor')
plt.title('Deadliest Actors in Hollywood (by Kill Count)')

# Display the kill count labels on each bar
for i, (actor, kill_count) in enumerate(zip(sorted_actors, sorted_kill_counts)):
    plt.text(kill_count + 20, i, str(kill_count), va='center', color='black')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Hypothetical data: number of Roman Emperors and whether they were assassinated
assassinated = 13
not_assassinated = 47

# Labels for the pie chart
labels = ['Assassinated', 'Not Assassinated']

# Data to be plotted
sizes = [assassinated, not_assassinated]
colors = ['red', 'green']  # Red for assassinated, green for not assassinated

# Create the pie chart
plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor': 'black'})

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Title of the chart
plt.title('Fraction of Roman Emperors Assassinated')

# Display the pie chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data for the years 2000-2009
years = np.arange(2000, 2010)
arcade_revenue = [800, 850, 880, 900, 950, 1000, 1050, 1100, 1150, 1200]  # in millions
cs_phds_awarded = [1200, 1300, 1350, 1400, 1450, 1500, 1600, 1700, 1750, 1800]  # number of PhDs

# Scatter plot
plt.figure(figsize=(8, 6))
scatter = plt.scatter(arcade_revenue, cs_phds_awarded, c=years, cmap='viridis', s=100, edgecolor='black')

# Labels
plt.xlabel('Arcade Revenue (in millions)')
plt.ylabel('Number of Computer Science PhDs Awarded')
plt.title('Relationship Between Arcade Revenue and Computer Science PhDs Awarded (2000-2009)')

# Add a colorbar to show the year values
plt.colorbar(scatter, label='Year')

# Display the plot
plt.show()
```
